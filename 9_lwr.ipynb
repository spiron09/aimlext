```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_train = np.array(list(range(3,33)) + [3.3,4.2])
X_train = X_train[:,np.newaxis]
Y_train = np.array([1,1,1,2,2,3,4,3,5,1,6,7,7,8,9,9,10,11,13,15,16,11,22,13,16,19,20,22,23,25,26,27])
X_test = np.array([i/10 for i in range(400)])
X_test = X_test[:,np.newaxis]
Y_test = []
tou=0.8
```

```python
count = 0
for i in range(len(X_test)):
    
    diff = X_train - X_test[i]
    squared = np.sum(diff**2, axis=1)
    weights = np.exp(-squared/2*tou**2)

    weight_matrix = np.diag(weights)

    XTX = X_train.T.dot(weight_matrix).dot(X_train)
    XTY = X_train.T.dot(weight_matrix).dot(Y_train)
    parameters = np.linalg.inv(XTX).dot(XTY)

    prediction = X_test[i].dot(parameters)

    Y_test.append(prediction)

plt.plot(X_train.squeeze(), Y_train, 'o', label="Training Data")
plt.plot(X_test.squeeze(), Y_test, '*', label="Test Data")
plt.legend()
plt.show()
```
